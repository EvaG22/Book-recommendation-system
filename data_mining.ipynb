```python
def get_books_cover(DIR_PATH, books, bookIds):
        """
        Retrieves books covers to a img/ directory
        Will work on existing books class attribute, so a GoodReads list should be scraped or a books list loaded
        (csv_to_books) before use.
        :return: None
        """
        img_dir = "img"
        check_folder = os.path.isdir(img_dir)

        # If folder doesn't exist, then create it.
        if not check_folder:
            os.makedirs( DIR_PATH + img_dir)
            print("Creating folder: ", img_dir)

        else:
            print(img_dir, "folder already exists, saving images to folder.")

        # Download covers
        for (book, bookId) in zip(books, bookIds):            
                urllib.request.urlretrieve(
                     book, "img/" + bookId  + ".jpg"
                )
                # Set a respectful wait time
                time.sleep(2)
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer

# Filter only books where language is English
df_english = df[df['language'] == 'English']

# Convert 'description' column to string data type
df_english.loc[:, 'description'] = df_english['description'].astype(str)

# Create an instance of TfidfVectorizer with unigrams and bigrams and remove stop words
tfidf = TfidfVectorizer(ngram_range=(1, 2), stop_words='english')

# Fit and transform 'description' column using the TfidfVectorizer instance
tfidf_matrix = tfidf.fit_transform(df_english['description'])

# Get the BookId and TF-IDF matrix as output
book_ids = df_english['bookId'].values

# print(tfidf_matrix.shape)
# print(book_ids.shape)

```

```python
from sklearn.metrics.pairwise import cosine_similarity
import numpy as np

def recommend(item_id, num):
    # Find the row index of the given book ID in the TF-IDF matrix
    idx = np.where(book_ids == item_id)[0][0]

    # Calculate the cosine similarity between the given book and all the others
    sim_scores = cosine_similarity(tfidf_matrix[idx], tfidf_matrix)

    # Get the indices of the books with the highest similarity scores
    sim_scores_idx = sim_scores.argsort()[0][-num-1:-1][::-1]

    # Create a dictionary to store the recommended books
    recommended_books = {}

    # Get the recommended books and their similarity scores
    for i in sim_scores_idx:
        recommended_books[df_english.iloc[i]['title']] = {'description': df_english.iloc[i]['description'], 'score': sim_scores[0][i]}

    # Print the recommended books
    print(f"Recommending {num} books similar to: {df_english.iloc[idx]['title']}")
    print("-" * 55)
    for i, (book, details) in enumerate(recommended_books.items()):
        print(f"Recommended {i+1}: {book}")
        print(f"Description: {details['description']}")
        print(f"(score: {details['score']})")
        print()
```

```python
import re
import string

def dlean_text(raw_text):
  text = re.sub(r'[^\x00-\x7F]+', '', raw_text)
  text = text.strip()
  text = ''.join(char for char in text if char not in string.punctuation)
  text = text.lower()

  return text
```

```python
#takes almost 2 minutes

import nltk
from nltk.corpus import stopwords
from nltk.stem import WordNetLemmatizer
import string
from gensim.models import Word2Vec
import pickle

# Load stopwords and initialize lemmatizer
nltk.download('stopwords')
nltk.download('wordnet')
stopwords = set(stopwords.words('english'))
lemmatizer = WordNetLemmatizer()

# Function to clean the text
def clean_text(text):
    if isinstance(text, str):
        # Remove punctuation
        text = text.translate(str.maketrans('', '', string.punctuation))
        # Convert to lowercase
        text = text.lower()
        # Tokenize text
        tokens = text.split()
        # Remove stopwords and lemmatize tokens
        tokens = [lemmatizer.lemmatize(token) for token in tokens if token not in stopwords]
        return tokens
    else:
        return []

# Apply the cleaning function to the 'description' column
df['description_tokens'] = df['description'].apply(clean_text)

# Train Word2Vec model on the description tokens
model = Word2Vec(df['description_tokens'], vector_size=200, window=5, min_count=1, workers=4)

# Save the Word2Vec model as a pickle file
model.save('word2vec_model.pkl')

# Calculate the average embedding for each description
df['description_embeddings'] = df['description_tokens'].apply(lambda tokens: 
                                                              model.wv[tokens].mean(axis=0)
                                                              if tokens else None)

# Save the dataframe with the description embeddings as a pickle file
df.to_pickle('description_embeddings.pkl')
```

```python
def compute_average_embedding(text):
    tokens = text.split()
    embeddings = [model.wv[token] for token in tokens if token in model.wv.vocab]

    if embeddings:
        avg_embedding = np.mean(embeddings, axis=0)
        return avg_embedding
    else:
        return np.zeros(model.vector_size)
```

```python
def compute_average_embedding(text):
    text = str(text)
    tokens = text.split()
    embeddings = [model.wv.get_vector(token) for token in tokens if token in model.wv.key_to_index]

    if embeddings:
        avg_embedding = np.mean(embeddings, axis=0)
        return avg_embedding
    else:
        return np.zeros(model.vector_size)
```

```python
selected_books['embedding'] = selected_books['description'].apply(compute_average_embedding)
selected_books
```

```python
from sklearn.model_selection import train_test_split

x = selected_books['embedding'].tolist()
y = selected_books['genreSingle'].tolist()

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)
```

```python
from sklearn.naive_bayes import MultinomialNB
from sklearn.svm import SVC
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import GridSearchCV
from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler()
x_train = scaler.fit_transform(x_train)
x_test = scaler.transform(x_test)

# print(x_test)
# print(x_train)

# Convert lists to NumPy arrays
x_train = np.array(x_train)
x_test = np.array(x_test)

# Create and train the Naive Bayes classifier
naive_bayes = MultinomialNB()
naive_bayes.fit(x_train, y_train)

# Create and train the Support Vector Machines (SVM) classifier
parameters_svm = {'C': [1], 'gamma': [0.1], 'kernel': ['rbf']}
svm = SVC()
# Split the data into a smaller training set and a validation set
x_train_small, _, y_train_small, _ = train_test_split(x_train, y_train, train_size=0.2, random_state=42)
# Perform grid search on the smaller training set
grid_search_svm = GridSearchCV(svm, parameters_svm)
grid_search_svm.fit(x_train_small, y_train_small)
best_svm = grid_search_svm.best_estimator_
print("Best SVM parameters:", grid_search_svm.best_params_)

# Create and train the Random Forests classifier
random_forest = RandomForestClassifier()
random_forest.fit(x_train, y_train)


# Perform predictions
nb_predictions = naive_bayes.predict(x_test)
svm_predictions = best_svm.predict(x_test)
rf_predictions = random_forest.predict(x_test)
```

```python
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score

def print_evaluation_scores(y_true, y_pred):
    accuracy = accuracy_score(y_true, y_pred)
    precision = precision_score(y_true, y_pred, average='macro')
    recall = recall_score(y_true, y_pred, average='macro')
    f1 = f1_score(y_true, y_pred, average='macro')

    print("Accuracy:", accuracy)
    print("Precision:", precision)
    print("Recall:", recall)
    print("F1 Score:", f1)
```
